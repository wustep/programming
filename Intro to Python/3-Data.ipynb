```python
import csv

with open('data/KHC.csv', 'r') as f:
    reader = csv.reader(f)
    data = list(reader)
    
print(data)
```

```python
# remove header

data = data[1:]

total = 0.0
count = 0.0
for i in range(len(data)):
    total += float(data[i][1])
    count += 1

print(total / count)
```

```python
%matplotlib inline

import matplotlib 
import matplotlib.pyplot as plt

data_open_price = [float(x[1]) for x in data]
data_volume = [float(x[6]) for x in data]

plt.plot(data_volume, data_open_price)
plt.ylabel('open price')
plt.xlabel('volume')
plt.show()
```
